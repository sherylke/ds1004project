### Housing price and Crime frequency

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
zipcode = pd.read_table('col5_compared_to_zipcode.out',header=-1)

zipcode.columns = ["zipcode","year","month",'weeknum',"weekday","freq"]
#zipcode["zipcode"] =  zipcode["zipcode"]
zipcode["month"] =  zipcode["month"].astype(str)
zipcode["year"] =  zipcode["year"].astype(str)
zipcode["weeknum"] =  zipcode["weeknum"].astype(str)
zipcode["weekday"] =  zipcode["weekday"].astype(str)

zip_grouped = zipcode.groupby(["zipcode"],sort=True).sum()
```

```python
house = pd.read_csv('Zip_Zhvi_AllHomes.csv')
```

```python
ny_house = house[house["Metro"]=="New York"]
ny_house_time = ny_house[ny_house.columns[124:-15]]

nyc_house = pd.DataFrame(
    {'zipcode': ny_house["RegionName"],
     'avg_price': ny_house_time.mean(axis =1)
    })        
```

```python
zip_code_list = list()
for i in zip_grouped.index:
    zip_code_list.append(int(i))
zip_grouped["zipcode"] = zip_code_list 
```

```python
zip_and_price = pd.merge(zip_grouped, nyc_house, on='zipcode', how='left')
```

```python
zip_and_price = zip_and_price.dropna(axis=0, how='any')
zip_and_price.to_csv("zipcode and price.csv")
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(111)
ax1.plot(list(zip_and_price["freq"]),color = "red",label = "Number of Crimes")
#ax1.set_xticks(list(grouped.index))
ax1.set_ylabel("Total Number of Crimes")
plt.legend(bbox_to_anchor=(1.05, 1), loc=3, borderaxespad=0.)


ax2 = ax1.twinx()
ax2.plot(list(zip_and_price["avg_price"]),color = "blue",label = "Avg House Price")
#ax2.set_xticks(list(grouped.index))
ax2.set_ylabel("Avg Price")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
