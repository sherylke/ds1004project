## Some Trends

### 1. Number of Compliants by Year

Code: ***col5_report_year.py*** This code returns number of compliants by year. 

```python
df = pd.read_table('col5_report_year.out',header=-1)
df = df.sort([0], ascending=True)
plt.plot(list(df[0]),list(df[1]),linestyle='--', marker='o')
plt.title("Number of Compliant by Year")
plt.xlabel("Year")
plt.ylabel("Number of Compliant")
plt.show()
```

**Trend:** Overall, the crime decreased year over year. 

### 2. Number of compliants by year and by month

Code: ***col5_report_year_month.py*** This code returns number of compliants by each year and each month. 

```python
fig = plt.figure(figsize=(10, 8))
df = pd.read_table('col5_report_year_month.out',header=-1)
df = df.sort([0,1], ascending=True)
for i in df[0].unique():
    plt.plot(list(df[df[0]==i][1]),list(df[df[0]==i][2]),linestyle='--', marker='o',label = i)
plt.xticks(list(range(1,13)))
plt.legend(loc="upper left")
plt.title("Number of compliants by year and by month")
plt.show()
```

It seems that Jan and Feb in 2015 has some abnormal number of compliants compared to the other month.

I also plot a boxplot and clearly Feburay has the lowest number of compliants on average (partialy is becasue of shorter month). It seems there is an outliar in May and December.  

```python
fig = plt.figure(figsize=(10, 8))
month = list()
for i in (list(range(1,13))):
    month.append(df[df[1]==i][2])

plt.boxplot(month)
plt.title("Box Plot by Month")
plt.show()
```

### 3. Number of compliants by year and by week

Code: ***col5_report_year_week.py*** This code provides number of complaints by year and by week.  

```python
fig = plt.figure(figsize=(15, 15))
df = pd.read_table('col5_report_year_week.out',header=-1)
df = df.sort([0,1], ascending=True)
for i in df[0].unique():
    plt.plot(list(df[df[0]==i][1]),list(df[df[0]==i][2]),linestyle='--', marker='o',label = i)
plt.legend(loc="upper left")
plt.xticks(list(df[1].unique()))
plt.show()
```

After plotting by week, it is clear that the week 44 in 2012 has an abnormal trend. It is probably because of hurricane sandy happened in that period.

### 4. Number of compliants by day of week

Code: ***col5_report_year_week_day.py*** This code provides number of compliants by year and weeknum in each day of the week.



```python
fig = plt.figure(figsize=(12, 10))
df = pd.read_table('col5_report_year_week_day.out',header=-1)
df = df.sort([0,1], ascending=True)
dow = list()
for i in (list(range(1,8))):
    dow.append(df[df[2]==i][3])
plt.boxplot(dow)
plt.show()
```

By looking at the boxplot, Tue-Fri is the peak and weekend has less compliants. Also, weekdays have more outliars. 

### 5. Number of compliants by hour

Code: ***col1and2_hour.py***. This code is to check number of compliants by hour. 

```python
fig = plt.figure(figsize=(12, 10))
df = pd.read_table('col1and2_hour.out',header=-1)
df = df.sort([0,1,2,3], ascending=True)
dow = list()
for i in (list(range(0,24))):
    dow.append(df[df[3]==i][4])
plt.boxplot(dow)
plt.xticks(list(range(1,25)),list(range(0,24)))
plt.show()
```

Number of complaint is gradually increasing after 7 am and then start decreasing after midnight. However, there are some very high abnormal outliars happened between 12 - 1 am. 

```python
dfzero = df[df[3]==0]
dfoutliar = dfzero[dfzero[4]>300]
dfoutliar
```

The outliars are usually between 12 and 1 am on the new year. It makes sense since the NYE celebrations.

### 6. Number of compliants by length of compliant 

Code: ***length_of_compliant.py***. The code provides number of compliants by the length of complaint. When the compliant has no to_date and to_time, the length will be counted as **0**. 

```python
fig = plt.figure(figsize=(12, 10))
df = pd.read_table('length_of_compliant.out',header=-1)
df = df.sort([0], ascending=True)

plt.plot(list(df[0]),list(df[1]),linestyle='-')
plt.title("Number of compliant by length of compliant (Log-Log)")
plt.xlabel("Length of compliant")
plt.ylabel("Number of Compliant")
plt.xscale('log')
plt.yscale('log')
plt.show()
```

Overall,number of compliants decays over the length of the compliant.
