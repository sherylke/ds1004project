# Some Data Distributions

### Column 11 - Offense Level

```python
data = pd.read_table('column11_data_quality.out', header = -1)
```

```python
stats = {}
for level in ['FELONY','MISDEMEANOR','VIOLATION']:
    stats[level] = len(data[data[0]==level])
```

```python
plt.bar(range(len(stats)),stats.values(),align='center')
plt.xticks(range(len(stats)), stats.keys())

plt.show()
```

### Column 13 - Borough Names

```python
data = pd.read_table('column13_data_quality.out', header = -1)
```

```python
stats = {}
for boro in ['BRONX','BROOKLYN','MANHATTAN','QUEENS','STATEN ISLAND']:
    stats[boro] = len(data[data[0]==boro])
```

```python
plt.bar(range(len(stats)),stats.values(),align='center')
plt.xticks(range(len(stats)), stats.keys(), rotation=17)

plt.show()
```
